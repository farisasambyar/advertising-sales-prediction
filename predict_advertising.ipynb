# Prediksi penjualan berasarkan biaya pengiklanan TV, Radio, Koran

## üìå Import Library

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

pandas ‚Üí untuk mengolah data dalam bentuk tabel (DataFrame).

numpy ‚Üí untuk operasi numerik, array.

train_test_split ‚Üí untuk membagi dataset menjadi data training dan testing.

LinearRegression ‚Üí model regresi linear untuk prediksi.

## üìå Baca Dataset

```python
file_url = "https://drive.google.com/uc?export=download&id=1wi4TTpTJXRyGMXj_em-AL78U7GgtLXrq"

data = pd.read_csv(file_url)

print(data.head())
```

Membaca file advertising.csv

head() menampilkan 5 baris pertama dataset.

Dataset ini  berisi kolom:
| TV | Radio | Newspaper | Sales |

Tujuannya: memprediksi Sales berdasarkan budget iklan TV, Radio, dan koran.

## üîß Membuat Dataset Latihan & Testing

```python
x = np.array(data.drop(["Total Penjualan"], axis=1))
y = np.array(data["Total Penjualan"])
```

x = semua fitur (TV, Radio, Newspaper)

y = target (Sales)

## ‚úÇ Split Data

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y,
                                                test_size=0.2,
                                                random_state=42)
```

Membagi dataset menjadi:

80% training

20% testing

random_state=42 agar hasil selalu sama saat di-run ulang.

## üß† Training Model



```python
model = LinearRegression()
model.fit(xtrain, ytrain)
```

Membuat model regresi linear.

.fit() melatih model menggunakan data training.

## üìà Evaluasi Model

```python
print(model.score(xtest, ytest))
```

Menampilkan nilai R-Squared (akurasi model prediksi).

Semakin mendekati 1, semakin baik.

## üîÆ Memprediksi Data Baru



```python
features = np.array([[230.1, 37.8, 69.2]])

hasil_prediksi = model.predict(features)

print(f"Hasil Prediksi adalah : {hasil_prediksi} ")
```

Memasukkan nilai iklan baru: TV=230.1, Radio=37.8, Koran=69.2

Model memberikan prediksi penjualan.
